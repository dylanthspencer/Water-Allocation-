```python
#modeling using inflow and outflow data
P= 8264256                #lake powell initial water volume
M= 8928903               #lake mead initial
Y_0 = 2025

P_min= 4000000        # powell (glen canyon dam) min water (acre-feat) for electricity generation
M_min= 6184000        # mead (hoover dam)

D_az=  2520000           #water demand in AF
D_ca=  4400000
D_wy=  504000
D_nm=  546098
D_co=  2120000
M_needs=   1500000      #mexico needs
D_total=D_az+D_ca+D_wy+D_nm+D_co+M_needs

state_demands={
    'AZ': D_az,
    'CA': D_ca,
    'WY': D_wy,
    'NM': D_nm,
    'CO': D_co,
    'Mexico': M_needs
}
distribution_over_time = {state: [] for state in state_demands.keys()}
years=list(range(1980,2026))

for i in range(46):
  for j in range (0,365):
    P_in = inflowmodelP.coef_[0][0]*(Y_0+i+(j/365))+inflowmodelP.intercept_[0] #lake powell inflow modeled
    P_out = 0#outflowmodelP.coef_[0][0]*(Y_0+i+(j/365))+outflowmodelP.intercept_[0] #lake powell outflow modeled
    M_in = inflowmodelM.coef_[0][0]*(Y_0+i+(j/365))+inflowmodelM.intercept_[0]
    M_out= 0#outflowmodelM.coef_[0][0]*(Y_0+i+(j/365))+outflowmodelM.intercept_[0]

    P+=P_in-P_out
    M+=M_in-M_out
  P = 0.9683127665*P
  M = 0.9408520732*P
  excess=P+M-D_total-M_min-P_min
  if excess>0:
    print("All needs met. Gulf Excess: " + str(excess))
    P-=D_total*(P-P_min)/(P+M-M_min-P_min)
    M-=D_total*(M-M_min)/(P+M-M_min-P_min)
    for state,demand in state_demands.items():
        distribution_over_time[state].append(demand)
  else:
    supply=max(P-P_min,0)+max(M-M_min,0)
    P=P_min
    M=M_min
    for state,demand in state_demands.items():
        state_distribution= demand*supply/D_total
        distribution_over_time[state].append(state_distribution)


plt.figure(figsize=(10, 6))
for state, values in distribution_over_time.items():
  plt.plot(years, values, label=state, marker="o")

plt.xlabel("Year")
plt.ylabel("Water Distribution (AF)")
plt.title("Water Distribution to Each State Over Time")
plt.legend()
plt.grid(True)
plt.show()
print(D_nm/D_total)




```
